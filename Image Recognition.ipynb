```python
import pandas as pd
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

## Goal
The goal of this notebook is that given an image of a single card let's determine what card it is. 

Feature extraction is typically used in combination with a neural network for classification. This is because processing a while image is pretty intensive and feature extraction can be seen like compressing the image into its important parts. There are also non-neural network classification methods such as NearestNeighbors and Support Vector MAchines (SVMs) but whether we use them depends on the final accuracy and speed

```python
im1 = cv2.imread("Images/CHUN-LI 001.png", 0)
#plt.figure(figsize=(20,20))
print(im1.shape)
plt.imshow(im1)
plt.show()
```

```python

#ORB feature detector
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(im1, None)
print(des1[0])
```
